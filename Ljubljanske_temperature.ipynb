```python
pod_df = pd.DataFrame({"indeks":[i for i in range(35)], "stolpec": [1 for i in range(35)]})
pod_df
```

```python
ATA = (pod_df.T).dot(pod_df)
```

```python
VR = (pod_df.T).dot(januar_df["TEMPERATURA"])
```

```python
VR
```

```python
import numpy
A_array = ATA.to_numpy()
b_array = VR.to_numpy()

# ReÅ¡imo linearni sistem Ax = b
x_array = numpy.linalg.solve(A_array, b_array)

x_array
```

```python
x_values = [i for i in range(35)]
y_values = [(x_array[0]*i+x_array[1]) for i in range(35)]
plt.plot(x_values,y_values,c='r')
plt.scatter(x_values, januar_df["TEMPERATURA"])

plt.show()
```
